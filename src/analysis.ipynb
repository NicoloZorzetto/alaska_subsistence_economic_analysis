## Aggregating for boroughs
Since most data like unemployment, taxes and income are census area-level, I will aggregate the rest to that.

```python
import pandas as pd
import numpy as np

# Convert numeric columns safely
numeric_cols = [
    "adj_households_in_community", "sqMiLand", "sqMiWater", 
    "using", "trying", "harvesting", "giving", "receving", 
    "units_harvested", "pounds_harvested", "units_harvested_by_sample", 
    "lbs_harvested_percapita", "lbs_harvested_by_sample", "lbs_used_percapita", 
    "perc_contribution_to_harvest"
]
an_df.loc[:, numeric_cols] = an_df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# Automatically detect borough-level variables (those that are NOT in numeric_cols + ["commname"])
borough_columns = [col for col in an_df.columns if col not in numeric_cols + ["commname"]]

# Define **weighted aggregation** for community-level data
def weighted_mean(series, weights):
    return np.average(series, weights=weights) if not series.isna().all() else np.nan

agg_funcs = {
    "adj_households_in_community": "sum",
    "sqMiLand": "sum",
    "sqMiWater": "sum",
    "using": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "trying": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "harvesting": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "giving": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "receving": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "units_harvested": "sum",
    "pounds_harvested": "sum",
    "units_harvested_by_sample": "sum",
    "lbs_harvested_percapita": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "lbs_harvested_by_sample": "sum",
    "lbs_used_percapita": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
    "perc_contribution_to_harvest": lambda x: weighted_mean(x, an_df.loc[x.index, "adj_households_in_community"]),
}

# Group only **community-level** data
grouped = an_df.groupby(["BoroughCensusArea", "year", "resource"]).agg(agg_funcs).reset_index()

# Retrieve **borough-level** variables (they are constant per group)
area_an_df = grouped.merge(
    an_df[borough_columns].drop_duplicates(["BoroughCensusArea", "year", "resource"]),
    on=["BoroughCensusArea", "year", "resource"],
    how="left"
)

# **Ensure `relying_on_substinence` stays intact**
area_an_df["relying_on_substinence"].fillna("NO", inplace=True)

# Recompute population density
area_an_df["TotsqMi"] = area_an_df["sqMiLand"] + area_an_df["sqMiWater"]
area_an_df = area_an_df[area_an_df["TotsqMi"] > 0]
area_an_df["pop_density"] = area_an_df["n_total_population(ACS)"] / area_an_df["TotsqMi"]

# **Final check**
print("Columns in final dataset:", area_an_df.columns)  # Should include all borough-level variables
print("Final unique values for `relying_on_substinence`:", area_an_df["relying_on_substinence"].unique())  # Should be ['YES', 'NO']
```

```python
# Fill missing `relying_on_substinence`
area_an_df["relying_on_substinence"].fillna("NO", inplace=True)

# Drop NaNs only in **critical columns**, not all
critical_cols = ["harvesting", "pop_density"]  # Adjust as needed
area_an_df = area_an_df.dropna(subset=critical_cols).reset_index(drop=True)

# Final check
print("Final unique values for `relying_on_substinence`:", area_an_df["relying_on_substinence"].unique())  
#  Should be ['YES', 'NO']
```

```python
# Investigating collinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import numpy as np

# Select independent variables
X = area_an_df[["local_tax_burden", "federal_tax_burden", "unemployment_rate", "AGI", "pop_density"]]

# Ensure no missing or infinite values
X = X.replace([np.inf, -np.inf], np.nan)  # Convert inf values to NaN
X = X.dropna()  # Drop rows with NaN values

# Add intercept
X = sm.add_constant(X)

# Compute VIF
vif_df = pd.DataFrame({
    "Variable": X.columns,
    "VIF": [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
})

# Display results
print(vif_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pygam import LinearGAM, s

# Select predictors
X = borough_df[[
    "p_households_below_poverty", 
    "p_population_native", 
    "local_tax_burden", 
    "federal_tax_burden", 
    "unemployment_rate", 
    "AGI_per_household"
]]
y = borough_df["trying_harvesting_pca"]

# Fit GAM model
gam = LinearGAM(
    s(0) + s(1) + s(2) + s(3) + s(4) + s(5)
).fit(X, y)

# Print summary
print(gam.summary())

# Plot Partial Dependence
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, ax in enumerate(axs.flatten()):
    XX = np.linspace(X.iloc[:, i].min(), X.iloc[:, i].max(), 100).reshape(-1, 1)
    pdep, confi = gam.partial_dependence(i, width=0.95)
    
    # Transpose confi to match expected shape
    confi = confi.T  # This ensures confi is (2, 100) for plotting

    ax.plot(XX, pdep, label="Effect")
    ax.fill_between(XX.flatten(), confi[0, :], confi[1, :], alpha=0.3, label="95% CI")
    ax.set_title(f"Effect of {X.columns[i]}")
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import statsmodels.formula.api as smf
from scipy.stats import chi2

# Fit the fixed-effects model (OLS)
fixed_model = smf.ols(
    "sharing_pca ~ p_households_below_poverty + lbs_harvested_percapita + lbs_harvested_by_sample + local_tax_burden + social_security_benefits + AGI_per_household + pop_density + employed_per_household",
    data=borough_df
).fit()

# Fit the mixed-effects model (random intercept for BoroughCensusArea)
mixed_model = smf.mixedlm(
    "sharing_pca ~ p_households_below_poverty + lbs_harvested_percapita + lbs_harvested_by_sample + local_tax_burden + social_security_benefits + AGI_per_household + pop_density + employed_per_household",
    data=borough_df, groups=borough_df["BoroughCensusArea"]
).fit()

# Likelihood Ratio Test (LRT)
lrt_stat = 2 * (mixed_model.llf - fixed_model.llf)  # 2 * log-likelihood difference
df_diff = fixed_model.df_resid - mixed_model.df_resid  # Difference in degrees of freedom
p_value = chi2.sf(lrt_stat, df_diff)  # Compute p-value

# Print results
print(f"Likelihood Ratio Test Statistic: {lrt_stat:.4f}")
print(f"Degrees of Freedom Difference: {df_diff}")
print(f"P-Value: {p_value:.4f}")

# Interpretation:
if p_value < 0.05:
    print("Significant difference: Mixed model is preferred.")
else:
    print("No significant difference: Fixed model is sufficient.")
```
